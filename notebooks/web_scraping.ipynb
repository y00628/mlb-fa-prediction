```python
import requests
import pandas as pd
```

```python
webpage = requests.get('https://www.fangraphs.com/api/tools/free-agent-tracker/v2/free-agents?season=2025')
```

```python
data = pd.DataFrame(webpage.json())
```

```python
years = [2020, 2021, 2022, 2023, 2024, 2025]
for year in years:
    webpage = requests.get(f'https://www.fangraphs.com/api/tools/free-agent-tracker/v2/free-agents?season={str(year)}')
    data = pd.DataFrame(webpage.json())
    data.to_csv(f'../data/{year}.csv', index=False)
    print(f'{year} done')
```
